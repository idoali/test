# Data Preprocessing

```python
# Input data
import pandas as pd

data = pd.read_csv("data_input/trending.csv")
data.head()
```

```python
# Tampilkan kolom
data.columns
```

```python
# Buat data baru berdasarkan kolom yang kita inginkan saja
data = data[['title', 'channel_name', 'category_id', 'view', 'like', 'dislike', 'comment', 'trending_time']]
```

```python
data.head()
```

```python
# Ubah category_id jadi category_name. Tapi pertama-tama kita harus meng-import category_names terlebih dahulu 
import json

with open("data_input/category.json", "rb") as f:
    categories = json.load(f)
```

```python
categories = pd.json_normalize(categories, record_path="items")
categories_dict = {idx:title for idx, title in zip(categories['id'], categories['snippet.title'])}
```

```python
data['category'] = [categories_dict[str(x)] if str(x) in list(categories_dict.keys()) else "Unknown" for x in data['category_id']]
```

```python
data
```

```python
# Lihat semua kategori dan jumlah mereka
data['category'].value_counts()
```

```python
# Cek jika ada nilai yang bernilai NA
data.isnull().sum()
```

```python
# Mengubah semua nilai bernilai NA menjadi 0
data = data.fillna(0)
```

```python
# Mencari nilai trending_date dan menghapus trending_time
trending = pd.to_datetime(data['trending_time'])
data['trending_date'] = trending.dt.date
data = data.drop(['trending_time'], axis=1)
```
